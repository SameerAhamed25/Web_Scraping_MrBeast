### **Loading Web Pages with 'request'**

The requests module allows you to send HTTP requests using Python.

The HTTP request returns a Response Object with all the response data (content, encoding, status, and so on).


```python
import requests

res = requests.get('https://mrbeast.store/')

txt = res.text
status = res.status_code
print(txt, status)
```

### **Select DOM elements with BeautifulSoup methods.**

```python
import requests
from bs4 import BeautifulSoup

page = requests.get("https://mrbeast.store/")
soup = BeautifulSoup(page.content, 'html.parser')

all_a_tags = []

for element in soup.find_all('a'):
    all_a_tags.append(element.text.strip())

print(all_a_tags)
```

### **Elements to scrap**
in this case [Style]

```python
import requests
from bs4 import BeautifulSoup

# Make a request to your website
page = requests.get("https://mrbeast.store/")
soup = BeautifulSoup(page.content, 'html.parser')

# Create top_items as an empty list
top_items = []

# Extract and store in top_items according to instructions
products = soup.select('head')  # Adjust the selector
for elem in products:
    title = elem.select('script')[0].text  # Adjust the selector
    review_label = elem.select('style')[0].text  # Adjust the selector
    info = {
        "title": title.strip(),
        "review": review_label.strip()
    }
    top_items.append(info)

print(top_items)
```

### **Extracting Links**

```python
import requests
from bs4 import BeautifulSoup

page = requests.get("https://mrbeast.store/")
soup = BeautifulSoup(page.content, 'html.parser')

extracted_links = []

# Extract links from various elements
link_elements = soup.find_all(['a', 'link', 'area', 'img'])

# extract the 'href' attribute
for elem in link_elements:
    href = elem.get('href')
    if href:
        extracted_links.append(href)

print(extracted_links)
```

### **Generating a CSV file**

```python
import requests
from bs4 import BeautifulSoup
import csv

page = requests.get("https://mrbeast.store/")
soup = BeautifulSoup(page.content, 'html.parser')

extracted_links = []

# Extract links from various elements
link_elements = soup.find_all(['a', 'link', 'area', 'img'])

# extract the 'href' attribute
for elem in link_elements:
    href = elem.get('href')
    if href:
        extracted_links.append(href)

# Writing the extracted links to a CSV file
with open('extracted_links.csv', 'w', newline='') as csvfile:
    csv_writer = csv.writer(csvfile)
    csv_writer.writerow(['Link'])

    for link in extracted_links:
        csv_writer.writerow([link])

print("CSV file 'extracted_links.csv' has been created.")
```
